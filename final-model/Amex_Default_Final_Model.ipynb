```python
# imports

import os
import gc
import glob
import tqdm
import numpy as np
import pandas as pd
from supervised.automl import AutoML # mljar-supervised
```

```python
# load data

test_pkls = sorted(glob.glob('/content/drive/MyDrive/kaggle/amex-default-predict/test_data_*'))
```

```python
# load ml fitted model

automl = AutoML('/content/drive/MyDrive/kaggle/amex-default-predict/AutoML_1')
```

```python
# making preditions

submission_df = pd.DataFrame()

for i in range(len(test_pkls)):
  print(i)
  df = pd.read_pickle(f'/content/drive/MyDrive/kaggle/amex-default-predict/test_data_{i}.pkl')
  df_customer = pd.DataFrame(df.index, columns=['customer_ID'])

  pred = (automl.predict_proba(df)[:, 1].reshape(-1, ).astype(np.float32).round(2))
  df_predict = pd.DataFrame(pred, columns=['prediction'])

  final_df = pd.concat([df_customer, df_predict], axis=1)
  submission_df = submission_df.append(final_df)

  del pred, df, df_customer, df_predict, final_df
```

```python
# sabe submission file 

submission_df.to_csv('/content/drive/MyDrive/kaggle/amex-default-predict/submission_1.csv', index=False)
```
